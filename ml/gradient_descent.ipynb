```python
# Gradient Descent to find local minima of y = (x + 3)^2


# Function
def f(x):
    return (x + 3) ** 2


# Derivative
def df(x):
    return 2 * (x + 3)


# Parameters
x = 2  # starting point
lr = 0.1  # learning rate
epochs = 20  # number of iterations

for i in range(epochs):
    grad = df(x)
    x = x - lr * grad
    print(f"Iter {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")

print("\nLocal minima occurs at x ≈", round(x, 4))
```
